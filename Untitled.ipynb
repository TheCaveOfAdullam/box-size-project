```python
import os
```

```python
source_root = 'C:/Users/USER/Desktop/47.물류공간 예측 데이터/3.개방데이터/1.데이터/Validation/01.원천데이터'
label_root = 'C:/Users/USER/Desktop/47.물류공간 예측 데이터/3.개방데이터/1.데이터/Validation/02.라벨링데이터'
```

```python
image_count = 0
label_count = 0
missing_labels = []
category_distribution = {}
```

```python
for dirpath, _, filenames in os.walk(source_root):
    for fname in filenames:
        if fname.endswith('.jpg'):
            image_count += 1
            # 상대 경로 추출
            relative_path = os.path.relpath(os.path.join(dirpath, fname), source_root)
            label_path = os.path.join(label_root, os.path.splitext(relative_path)[0] + '.json')
            
            # 카테고리 추정 (ex. 입고물품/가공식품/조미료 등)
            category_key = '/'.join(relative_path.split(os.sep)[:4])
            category_distribution[category_key] = category_distribution.get(category_key, 0) + 1

            if os.path.exists(label_path):
                label_count += 1
            else:
                missing_labels.append(relative_path)
```

```python
print(f"전체 이미지 수: {image_count}")
print(f"전체 라벨 수: {label_count}")
print(f"라벨 누락 이미지 수: {len(missing_labels)}")

print("\n카테고리별 이미지 수:")
for category, count in category_distribution.items():
    print(f" - {category}: {count}개")

if missing_labels:
    print("\n❌ 라벨이 누락된 이미지 예시:")
    for miss in missing_labels[:5]:
        print(" -", miss)
```
